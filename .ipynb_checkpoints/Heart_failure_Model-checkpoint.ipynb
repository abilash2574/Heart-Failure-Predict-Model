# Aim of this Notebook

In this notebook, we visualize the distribution of the variables in the [Heart Failure Prediction](https://www.kaggle.com/andrewmvd/heart-failure-clinical-data) dataset and explore their relationship with the target variable ```DEATH_EVENT```. Based on the result of this analysis, we define machine learning models to compare their performance on this dataset to predict the target variable using the most correlated subset of variables from the dataset.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
dataset = pd.read_csv('heart_failure_clinical_records_dataset.csv')
````

## Data Description

````python
dataset.sample(10)
````

### Variables in this dataset

1. Age: Age of the patient in years
2. Anaemia: Decrease of red blood cells or hemoglobin (0:Reduced or 1:Normal)
3. creatinine_phosphokinase: Level of the CPK enzyme in the blood (mcg/L)
4. Diabetes: If the patient has diabetes (0:No or 1:Yes)
5. ejection_fraction: Percentage of blood leaving the heart at each contraction (percentage)
6. high_blood_pressure: If the patient has hypertension (0:No or 1:Yes)
7. platelets: Platelets in the blood (kiloplatelets/mL)
8. serum_creatinine: Level of serum creatinine in the blood (mg/dL)
9. serum_sodium: Level of serum sodium in the blood (mEq/L)
10. sex: Biological sex of the patient (0:Female or 1:Male)
11. smoking: If the patient is a smoker (0:No or 1: Yes)
12. time: Follow-up period in days
13. death_event: If the patient survived till the end of follow-up period (0:No or 1:Yes )

### Missing values in dataset!

````python
dataset.isnull().sum()
````

## Feature Selection

### Using ExtraTreesClassifier

````python

plt.rcParams['figure.figsize']=15,6 
sns.set_style("darkgrid")

x = dataset.iloc[:, :-1]
y = dataset.iloc[:,-1]

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_) 
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(12).plot(kind='barh')
plt.show()
````

From the above graph we can see that ```time```, ```ejection_fraciton```, ```serum_creatinine```, ```age```, ```serum_sodium``` has the most feature importance on predicting the ```DEATH_EVENT```.

### Using Correlation Heat map 

````python
correlation = dataset.corr()
ax, fig = plt.subplots(figsize=(12,12))
sns.heatmap(correlation, vmin=-1, cmap='coolwarm', annot=True)
plt.show()
````

````python
correlation[abs(correlation['DEATH_EVENT']) > 0.1]['DEATH_EVENT']
````

As we can see here, we get the same features as above that has higher correlation with `DEATH_EVENT`
